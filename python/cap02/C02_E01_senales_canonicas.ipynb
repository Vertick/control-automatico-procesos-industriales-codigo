## 1. Importaciones

```python
import numpy as np
import matplotlib.pyplot as plt
import os
np.random.seed(0)
t = np.linspace(0, 10, 1000)
```

## 2. Cuatro señales canónicas y ruido

```python
escalon = np.where(t >= 1, 1.0, 0.0)
rampa   = np.maximum(0.0, t - 1)
sinus   = np.sin(2*np.pi*0.5*t)
ruido   = sinus + 0.2*np.random.randn(t.size)
```

## 3. Visualización

```python
fig, ax = plt.subplots(2, 2, figsize=(9, 5.2))
for a, s, n in zip(ax.flat,
                   [escalon, rampa, sinus, ruido],
                   ["Escalón", "Rampa", "Sinusoidal", "Sinusoidal + ruido"]):
    a.plot(t, s)
    a.set_title(n)
    a.grid(alpha=0.3)
    a.set_xlabel("t (s)")
fig.tight_layout()
out = '../../figures/cap02/fig_C02_F01_senales_canonicas.png'
os.makedirs(os.path.dirname(out), exist_ok=True)
fig.savefig(out, dpi=120)
plt.show()
```

## 4. Interpretación

Las cuatro señales canónicas son las entradas estándar para análisis de sistemas. Escalón modela cambios bruscos; rampa, evoluciones graduales; impulso (no mostrado, equivalente a una delta), excitación instantánea; sinusoidal, comportamiento en frecuencia. La sinusoidal con ruido ilustra cómo el filtrado del transmisor o de la acción derivativa del PID condiciona la calidad del lazo.
