## Ejemplo industrial

Sensor de caudal con curva no lineal q(h) = k·sqrt(h). Linealización en el punto de operación nominal.

## 1. Linealización por Taylor de q_out(h) = k·sqrt(h)

```python
import sympy as sp
h, k, h0 = sp.symbols("h k h0", positive=True)
q_out = k*sp.sqrt(h)
linear = q_out.subs(h, h0) + sp.diff(q_out, h).subs(h, h0)*(h - h0)
print("q_out lineal en torno a h0:")
print(sp.simplify(linear))
print()
print("Pendiente local:")
print(sp.diff(q_out, h).subs(h, h0))
```

## 2. Interpretación

La linealización aproxima la función no lineal `q_out = k·sqrt(h)` por una recta tangente en `h0`. La pendiente local `k/(2·sqrt(h0))` es la ganancia que el lazo verá alrededor de ese punto. La aproximación se degrada cuanto más se aleja `h` de `h0`: cuando esto ocurre, la sintonía optimizada para el régimen nominal se vuelve subóptima o incluso inestable, motivando técnicas de **gain scheduling** (Capítulo 13).
