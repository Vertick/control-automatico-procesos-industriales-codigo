## 1. FDT en serie y conversión a espacio de estados

```python
import numpy as np
import control as ct
import matplotlib.pyplot as plt
import os

G = ct.tf([1.0], [10.0, 1.0])      # planta de primer orden, tau=10
H = ct.tf([1.0], [0.5, 1.0])       # sensor con dinámica (tau=0.5)
L = G*H

print("Polos FDT:", ct.poles(L))
print("Ceros FDT:", ct.zeros(L))

ss = ct.tf2ss(L)
print()
print("Matriz A:"); print(ss.A)
print("Autovalores de A:", np.linalg.eigvals(ss.A))
```

## 2. Mapa de polos y ceros

```python
fig, ax = plt.subplots(figsize=(6, 5))
poles = ct.poles(L); zeros = ct.zeros(L)
ax.scatter(poles.real, poles.imag, s=80, marker='x', color='C0', label="Polos")
if len(zeros) > 0:
    ax.scatter(zeros.real, zeros.imag, s=80, marker='o', facecolors='none', edgecolors='C1', label="Ceros")
ax.axhline(0, color='gray', lw=0.5); ax.axvline(0, color='gray', lw=0.5)
ax.set_xlabel("Re(s)"); ax.set_ylabel("Im(s)")
ax.set_title("Mapa de polos y ceros de la FDT G·H")
ax.grid(alpha=0.3); ax.legend()
out = '../../figures/cap04/fig_C04_F02_polos_ceros.png'
os.makedirs(os.path.dirname(out), exist_ok=True)
fig.tight_layout(); fig.savefig(out, dpi=120); plt.show()
```

## 3. Interpretación

Los polos de la FDT y los autovalores de la matriz A coinciden, confirmando la equivalencia matemática entre FDT y espacio de estados (en ausencia de cancelaciones polo-cero). La FDT habla bien de polos, ceros y respuesta en frecuencia; el espacio de estados habla bien de estructura interna y resulta más natural para sistemas MIMO, MPC o control por observador.
