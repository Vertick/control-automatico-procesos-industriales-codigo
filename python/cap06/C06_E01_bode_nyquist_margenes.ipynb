## 1. Lazo abierto y márgenes

```python
import numpy as np
import control as ct
import matplotlib.pyplot as plt
import os

G = ct.tf([1.0], [10.0, 1.0])
PI = ct.tf([0.4*5.0, 0.4], [5.0, 0.0])
L = PI*G

gm, pm, wcg, wcp = ct.margin(L)
print(f"Margen ganancia (lineal): {gm:.3f}")
print(f"Margen ganancia (dB):     {20*np.log10(gm):.2f}" if gm>0 else "MG indefinido")
print(f"Margen fase (deg):        {pm:.2f}")
```

## 2. Diagrama de Bode

```python
fig = plt.figure(figsize=(8, 6))
ct.bode_plot(L, dB=True, deg=True)
out = '../../figures/cap06/fig_C06_F01_bode_margenes.png'
os.makedirs(os.path.dirname(out), exist_ok=True)
plt.savefig(out, dpi=120, bbox_inches='tight'); plt.show()
```

## 3. Diagrama de Nyquist

```python
fig = plt.figure(figsize=(6, 6))
ct.nyquist_plot(L)
plt.title("C06_E01 — Nyquist del lazo abierto")
out = '../../figures/cap06/fig_C06_F02_nyquist.png'
plt.savefig(out, dpi=120, bbox_inches='tight'); plt.show()
```

## 4. Interpretación

Margen de fase y de ganancia cuantifican robustez del lazo. Reglas industriales: MF ∈ [40°, 70°], MG ≥ 6 dB. Por debajo de MF = 30° el lazo es frágil. Regla práctica: MF ≈ 100·ζ. Aplicada aquí, el MF obtenido se traduce en un amortiguamiento equivalente coherente con el sobredisparo observado en simulación temporal.
