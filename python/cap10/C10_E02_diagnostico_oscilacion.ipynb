## 1. Generación de patrones de oscilación

```python
import numpy as np
import matplotlib.pyplot as plt
import os
np.random.seed(0)

t = np.linspace(0, 60, 1200)
# Oscilación senoidal por margen de fase pequeño
osc_margen = 1.0 + 0.3*np.sin(2*np.pi*0.3*t)
# Oscilación cuadrada / triangular (stiction de válvula)
osc_stiction = 1.0 + 0.3*np.sign(np.sin(2*np.pi*0.2*t))
# Lazo sano (escalón con ligero ruido)
osc_sano = np.where(t < 2, 0.0, 1.0 - np.exp(-(t-2)/4)) + 0.02*np.random.randn(t.size)
```

## 2. Visualización

```python
fig, axs = plt.subplots(3, 1, figsize=(8, 6), sharex=True)
axs[0].plot(t, osc_margen, label="Sinusoidal (margen pequeño)")
axs[0].set_ylabel("y"); axs[0].grid(alpha=0.3); axs[0].legend()
axs[1].plot(t, osc_stiction, color='C1', label="Cuadrada (stiction)")
axs[1].set_ylabel("y"); axs[1].grid(alpha=0.3); axs[1].legend()
axs[2].plot(t, osc_sano, color='C2', label="Lazo sano")
axs[2].set_xlabel("t (s)"); axs[2].set_ylabel("y"); axs[2].grid(alpha=0.3); axs[2].legend()
fig.suptitle("C10_E02 — Patrones de oscilación: diagnóstico clásico")
out = '../../figures/cap10/fig_C10_F02_diagnostico.png'
os.makedirs(os.path.dirname(out), exist_ok=True)
fig.tight_layout(); fig.savefig(out, dpi=120); plt.show()
```

## 3. Interpretación

La forma de la oscilación discrimina la causa: senoidal limpia → margen de fase pequeño (problema de sintonía); cuadrada o trapezoidal → stiction de válvula (problema mecánico, la sintonía no lo resuelve); lazo sano → respuesta amortiguada al escalón.
