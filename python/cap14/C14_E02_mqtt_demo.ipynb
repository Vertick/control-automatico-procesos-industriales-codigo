## 1. Demo conceptual de cliente MQTT (opcional)

Este notebook **no requiere broker activo**: muestra el patrón pub/sub. Si el lector quiere ejecutar realmente, debe levantar un broker local con `mosquitto -v` y descomentar las líneas marcadas.

```python
# Patrón didáctico — sin conexión real
def make_payload(lazo, pv, sp, modo):
    return {
        "lazo": lazo, "pv": pv, "sp": sp, "modo": modo,
        "ts": "2026-04-29T10:00:00Z"
    }

mensajes_publicados = []
def publicar(topic, payload):
    """Mock que registra en lista en lugar de enviar al broker."""
    mensajes_publicados.append((topic, payload))

# Loop de telemetría a 1 Hz (mock, 5 mensajes)
for i in range(5):
    p = make_payload("LIC-101", pv=75.0+0.1*i, sp=75.0, modo="AUTO")
    publicar("plant/area01/LIC-101/state", p)

for topic, payload in mensajes_publicados:
    print(topic, "→", payload)
```

## 3. Interpretación

MQTT es el protocolo pub/sub más usado en IIoT por su ligereza. Este demo muestra el patrón: pasarela industrial publica telemetría con frecuencia controlada; suscriptores (SCADA, gemelo digital, plataforma de mantenimiento predictivo) consumen los mensajes. **Limitaciones:** MQTT clásico no garantiza determinismo de tiempo; no usar para control directo en lazo crítico.
