## 1. Modelo de estados de un automatismo simple

```python
# Llenado-mezcla-vaciado: 4 estados
estados = {"E0":"Inicial","E1":"Llenado","E2":"Mezcla","E3":"Vaciado"}
transiciones = {
    ("E0","start"): "E1",
    ("E1","nivel_alto"): "E2",
    ("E2","tiempo_mezcla"): "E3",
    ("E3","nivel_bajo"): "E0",
}
estado = "E0"
trazas = [estado]
for ev in ["start","nivel_alto","tiempo_mezcla","nivel_bajo","start"]:
    nuevo = transiciones.get((estado, ev), estado)
    print(f"{estado} --[{ev}]--> {nuevo}")
    estado = nuevo
    trazas.append(estado)
print()
print("Traza final:", trazas)
```

## 2. Interpretación

GRAFCET / SFC modela automatismos secuenciales: pasos discretos con transiciones condicionales. **No confundir con control en lazo cerrado continuo:** GRAFCET vive en el mismo PLC que el bloque PID, pero su finalidad es coordinar la secuencia de operación, no regular variables continuas.
